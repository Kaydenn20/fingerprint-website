```python
import os
import glob
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from sklearn.metrics import classification_report
from sklearn.utils.class_weight import compute_class_weight
```

```python
file_path = 'C:/Users/rasdy/Desktop/SEM 5/FYP 2/fingerprint-based-blood-group-detection-main/dataset/dataset_blood_group'
# List all classes in the dataset
name_class = os.listdir(file_path)
print(name_class)
```

```python
# Get file paths for all images in the dataset
filepaths = list(glob.glob(file_path + '/**/*.*'))

# Example of printing a subset of file paths
print(filepaths[-6:-1])

# Extract labels from file paths
labels = list(map(lambda x: os.path.split(os.path.split(x)[0])[1], filepaths))
print(labels)
```

```python
# Create a DataFrame with file paths and labels
filepath = pd.Series(filepaths, name='Filepath').astype(str)
Labels = pd.Series(labels, name='Label')
data = pd.concat([filepath, Labels], axis=1)
data = data.sample(frac=1).reset_index(drop=True)
print(data.head(5))

# Visualize class distribution
counts = data.Label.value_counts()
sns.barplot(x=counts.index, y=counts)
plt.xlabel('Type')
plt.xticks(rotation=90)
plt.show()
```

```python
# Split data into training and testing sets
train, test = train_test_split(data, test_size=0.20, random_state=42)

# Visualize some images from the dataset
fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(10, 8), subplot_kw={'xticks': [], 'yticks': []})
for i, ax in enumerate(axes.flat):
    ax.imshow(plt.imread(data.Filepath[i]))
    ax.set_title(data.Label[i])
plt.tight_layout()
plt.show()
```

```python
# --- Start of Class Weighting Code ---

# 1. Get the unique class labels and their counts from the TRAINING data
#    Using unique from the training labels ensures we only consider classes present in training
#    NOTE: Changed train_df to train based on your split code
classes = np.unique(train['Label']) # <-- Use 'train' here and check column name ('Label' or 'labels'?)

# 2. Calculate class weights
#    ... (rest of compute_class_weight call is the same) ...
weights = compute_class_weight(
    class_weight='balanced',
    classes=classes,
    y=train['Label'] # <-- Use 'train' here and check column name
)

# 3. Create a dictionary mapping CLASS INDEX to WEIGHT
#    ... (rest of the code is the same, using label_to_weight) ...
label_to_weight = dict(zip(classes, weights))
print("Calculated label weights:", label_to_weight)

# --- End of Class Weighting Code ---
```

```python
# Set up ImageDataGenerator for training and validation data
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_generator = train_datagen.flow_from_dataframe(
    dataframe=train,
    x_col='Filepath',
    y_col='Label',
    target_size=(256, 256),
    class_mode='categorical',
    batch_size=32,
    shuffle=True,
    seed=42
)

valid_gen = train_datagen.flow_from_dataframe(
    dataframe=test,
    x_col='Filepath',
    y_col='Label',
    target_size=(256, 256),
    class_mode='categorical',
    batch_size=32,
    shuffle=False,
    seed=42
)
```

```python
# This cell MUST run AFTER the cell that creates train_generator

class_indices = train_generator.class_indices # <-- This line requires train_generator to exist
print("Generator class indices:", class_indices)

class_weight_dict = {class_indices[label]: weight for label, weight in label_to_weight.items()}
print("Final class_weight dictionary for Keras:", class_weight_dict)
```

```python
# Define the base pre-trained model
pretrained_model = ResNet50(
    input_shape=(256, 256, 3),
    include_top=False,
    weights='imagenet',
    pooling='avg'
)

pretrained_model.trainable = False

# Add layers for classification on top of ResNet50
x = Dense(128, activation="relu")(pretrained_model.output)
x = Dense(128, activation="relu")(x)
outputs = Dense(8, activation='softmax')(x)

model = Model(inputs=pretrained_model.input, outputs=outputs)

model.compile(
    optimizer="adam",
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
# --- Phase 1 Training ---
history_phase1 = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=20, # Consider increasing epochs after fixing imbalance
    validation_data=valid_gen, # <-- Change from test_generator to valid_gen
    validation_steps=valid_gen.samples // valid_gen.batch_size, # <-- Also change here
    class_weight=class_weight_dict # Add this back once you fix that step
)

# --- Phase 2 Training ---
# After unfreezing layers...
history_phase2 = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=20, # Consider increasing epochs
    validation_data=valid_gen, # <-- Change from test_generator to valid_gen
    validation_steps=valid_gen.samples // valid_gen.batch_size, # <-- Also change here
    class_weight=class_weight_dict, # Add this back once you fix that step
    # Add initial_epoch=history_phase1.epoch[-1] + 1 if continuing
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report

# Ensure history_phase1 and history_phase2 are defined from running your training cells
if 'history_phase1' in locals() and 'history_phase2' in locals():
    # Combine the history from both training phases
    combined_history = {}
    # Get all unique keys from both histories
    all_keys = set(history_phase1.history.keys()).union(history_phase2.history.keys())

    for key in all_keys:
        history1_data = history_phase1.history.get(key, []) # Get data, default to empty list if key missing
        history2_data = history_phase2.history.get(key, [])
        combined_history[key] = history1_data + history2_data # Concatenate lists

    # Plot training history: accuracy
    if 'accuracy' in combined_history and 'val_accuracy' in combined_history:
        pd.DataFrame(combined_history)[['accuracy', 'val_accuracy']].plot()
        plt.title("Model Accuracy Over Epochs")
        plt.xlabel("Epoch")
        plt.ylabel("Accuracy")
        plt.show()
    else:
        print("Accuracy or val_accuracy data not found in history for plotting.")


    # Plot training history: loss
    if 'loss' in combined_history and 'val_loss' in combined_history:
        pd.DataFrame(combined_history)[['loss', 'val_loss']].plot()
        plt.title("Model Loss Over Epochs")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.show()
    else:
        print("Loss or val_loss data not found in history for plotting.")

    # --- Classification Report ---
    # This section requires 'model', 'valid_gen', 'train_generator', and 'test'
    try:
        print("Evaluating model on validation data...")
        # Evaluate the model using valid_gen as used in your training code
        results = model.evaluate(valid_gen, verbose=0)
        print(f"Validation Loss: {results[0]:.5f}")
        print(f"Validation Accuracy: {results[1]*100:.2f}%")

        # Predict labels for the data in valid_gen
        print("Generating predictions on validation data...")
        valid_gen.reset() # Reset generator before predicting to ensure correct order
        # Calculate steps for prediction to handle potential last partial batch
        predict_steps = valid_gen.samples // valid_gen.batch_size + (valid_gen.samples % valid_gen.batch_size > 0)
        pred_probs = model.predict(valid_gen, steps=predict_steps)
        pred_labels = np.argmax(pred_probs, axis=1)

        # Map predicted numerical labels back to class names
        # Use class_indices from train_generator (or any generator fitted on the full label set)
        labels_map = train_generator.class_indices
        labels_map = dict((v, k) for k, v in labels_map.items())
        pred_class_names = [labels_map[k] for k in pred_labels]

        # Get true labels from the test DataFrame
        # Assuming valid_gen iterates over the 'test' DataFrame and shuffle=False
        y_true_class_names = test['Label'].iloc[:valid_gen.samples].tolist() # Take labels for the number of samples in valid_gen

        print("Generating classification report...")
        # Check if the number of true and predicted labels match before generating the report
        if len(y_true_class_names) == len(pred_class_names):
             print(classification_report(y_true_class_names, pred_class_names))
        else:
             print(f"Error: Mismatch between number of true labels ({len(y_true_class_names)}) and predicted labels ({len(pred_class_names)}).")
             print("Please ensure valid_gen was created from the 'test' DataFrame and shuffle=False.")


    except NameError as e:
        # This block is reached if model, valid_gen, train_generator, or test are not defined
        print(f"Error: Required object not found. {e}")
        print("Please ensure the preceding cells defining 'model', 'valid_gen', 'train_generator', and 'test' were run successfully.")
    except Exception as e:
        print(f"An error occurred during evaluation or classification report generation: {e}")


else:
    # This block is reached if history_phase1 or history_phase2 are not defined
    print("Error: Training history objects (history_phase1 or history_phase2) not found.")
    print("Please ensure the cells containing model.fit for both phases were run successfully.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.imagenet_utils import preprocess_input

# Load the pre-trained model
model = load_model('model.resnet.keras')

# Define the class labels
labels = {'A+': 0, 'A-': 1, 'AB+': 2, 'AB-': 3, 'B+': 4, 'B-': 5, 'O+': 6, 'O-': 7}
labels = dict((v, k) for k, v in labels.items())

# Example of loading a single image and making a prediction
img_path = 'C:/Users/rasdy/Desktop/SEM 5/FYP 2/fingerprint-based-blood-group-detection-main/dataset/dataset_blood_group/O+/cluster_6_13.BMP'

img = image.load_img(img_path, target_size=(256, 256))  # Load the image
x = image.img_to_array(img)  # Convert image to array
x = np.expand_dims(x, axis=0)  # Expand dimensions to match the model input
x = preprocess_input(x)  # Preprocess the image

result = model.predict(x)  # Predict using the model
predicted_class = np.argmax(result)  # Get the predicted class index

# Map the predicted class to the label
predicted_label = labels[predicted_class]
confidence = result[0][predicted_class] * 100  # Confidence level

# Display the image
plt.imshow(image.array_to_img(image.img_to_array(img) / 255.0))
plt.axis('off')  # Hide axes

# Display the prediction and confidence below the image
plt.title(f"Prediction: {predicted_label} with confidence {confidence:.2f}%")
plt.show()
```
